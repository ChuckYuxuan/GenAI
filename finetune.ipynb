```python
import os

def clean_directory(folder_path):
    # Iterate over the files in the directory
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)

        # Check if it's a file (as opposed to a directory)
        if os.path.isfile(file_path):
            # Remove images with specific naming pattern and not sequentially named
            if filename.startswith("000") and filename.endswith("_camera0.png"):
                os.remove(file_path)
                print(f"Removed: {file_path}")

            # Rename 'promts.txt' to 'prompt.txt'
            if filename == "promts.txt":
                new_file_path = os.path.join(folder_path, "prompt.txt")
                os.rename(file_path, new_file_path)
                print(f"Renamed: {file_path} to {new_file_path}")

def main(starting_directory):
    # Walk through all directories and subdirectories
    for root, dirs, files in os.walk(starting_directory):
        print(f"Cleaning directory: {root}")
        clean_directory(root)


    # Replace this with the path to the top-level directory containing your folders
main_directory = "/home/yz4397/Desktop/Generative_AI/Data/FIne_tune/front_cams"
main(main_directory)
```

```python
import os
from PIL import Image

def resize_images_in_folder(folder_path):
    # Specify the new size you want the images to be
    new_size = (512, 512)

    # List all files in the directory
    for filename in os.listdir(folder_path):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):  # Add or remove file types as needed
            file_path = os.path.join(folder_path, filename)
            
            # Open the image
            with Image.open(file_path) as img:
                # Resize the image
                img = img.resize(new_size, Image.ANTIALIAS)
                
                # Save the resized image back to the same location
                img.save(file_path)
                print(f"Resized {filename} to 512x512")

def main(starting_directory):
    # Walk through all directories and subdirectories
    for root, dirs, files in os.walk(starting_directory):
        print(f"Resizing images in directory: {root}")
        resize_images_in_folder(root)


    # Replace this with the path to the top-level directory containing your folders
main_directory = "/home/yz4397/Desktop/Generative_AI/Data/FIne_tune/front_cams"
main(main_directory)
```
